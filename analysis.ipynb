```python
import pandas as pd

df = pd.read_csv("/Users/tanveer/Downloads/sales_data.csv")
df.head()
```

```python
df.head()
```

```python
df["order_date"] = pd.to_datetime(df["order_date"])
df["month"] = df["order_date"].dt.to_period("M")
```

```python
monthly_revenue = df.groupby("month")["revenue"].sum()

monthly_revenue.plot(kind="bar")
```

```python
product_revenue = df.groupby("product_name")["revenue"].sum().sort_values(ascending=False)

product_revenue.plot(kind="bar")
```

```python
customer_revenue = df.groupby("customer_name")["revenue"].sum().sort_values(ascending=False)

customer_revenue.plot(kind="bar")
```

```python
country_revenue = df.groupby("country")["revenue"].sum()

country_revenue.plot(kind="bar")
```
